```python
import pandas as pd
import numpy as np
```

```python
data=pd.DataFrame({
    "name": ["Anna", "Dane", "David", "Kevin","Joe","Rose"],
    "weight": [45.2, 50.5, 62.1, 64.0, 59.0, np.nan],
    "height": [155, 160, 162, 170, 167, 159],
    "birth_year": [2000, 1997, 1996, 1987, 1998, 2003],
    "gender": ["female", "male", "male", "male", "male", "female"],
    "city": ["New York", "Chicago", "Austin", "New York", np.nan, "New York"]
, })
data
```

```python
name=pd.DataFrame({
    "nick_name":["Anna","David", "Dane", "Rose"],
    "full_name":["Anna Levis", "David John", "Dane Sebastian", "Rose Kimberly"]
})
name
```

```python
data=data.merge(name, how = "left", left_on = "name", right_on = "nick_name")
data
```

```python
data=data.drop(columns=["nick_name"])
data
```

```python
data["full_name"].fillna(data["name"], inplace=True)
data
```

```python
data.dropna()
```

```python
def multiply_by_2(h):
    return np.square(h/100)
data["BMI"] = data["height"].apply(multiply_by_2)
data
```

```python
data["BMI"]=data["weight"]/data["BMI"]
data
```

```python
def kategori(b):
        if b >= 40:
            return ('Obesity class 3')
        elif b > 35:
            return('Obesity class 2')
        elif b > 30:
            return('Obesity class 1')
        elif b > 25:
            return('Overweight')
        elif b>18.5:
            return ('Normal Weight')
        else:
            return('Underweight')
data["Kategori_BMI"]=data["BMI"].apply(kategori)
data
```
